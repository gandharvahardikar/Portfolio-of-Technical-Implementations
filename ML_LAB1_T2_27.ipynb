## Replacing Values in columns

```python

d = {'gas': 0, 'diesel': 1}

print('Before replacement:')
print(df['fueltype'].head())


df['fueltype'] = df['fueltype'].map(d)

print('After replacement:')
print(df['fueltype'].head())
```

# Data Visualization

## line plot

```python
sns.lineplot(x='enginesize', y='price', data=df)
```

```python
sns.barplot(x='fueltype', y='price', data=df)
```

## Bar Graph

```python
sns.catplot(x='fueltype', data=df, kind='count')
```

```python
sns.catplot(x='fueltype', data=df, kind='count', hue='aspiration')
```

## Histogram

```python
df['enginesize'].hist(bins=8)
```

## Pie Chart

```python
sizes = df['fueltype'].value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(sizes,
        labels=sizes.index,
        autopct='%1.2f%%',
        shadow=True)
plt.show()
```
